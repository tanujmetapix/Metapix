```python
import torch
import torchvision.transforms as transforms
from PIL import Image

# Load the pretrained model
model = torch.load('pretrained_models/enhancement_lol.pth')
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = model.to(device)
model.eval()

# Define the image transformation
transform = transforms.ToTensor()

# Function to enhance the image
def enhance_image(image_path):
    # Load and preprocess the image
    image = Image.open(image_path).convert('RGB')
    image = transform(image).unsqueeze(0).to(device)

    # Perform image enhancement using the MIRNetv2 model
    with torch.no_grad():
        enhanced_image = model(image)

    # Convert the enhanced image to PIL format
    enhanced_image = enhanced_image.clamp(0, 1).squeeze(0).cpu()
    enhanced_image = transforms.ToPILImage()(enhanced_image)

    return enhanced_image

# Path to the input image
input_image_path = 'path/to/input_image.jpg'

# Enhance the image
enhanced_image = enhance_image(input_image_path)

# Save the enhanced image
output_image_path = 'path/to/output_image.jpg'
enhanced_image.save(output_image_path)

print("Image enhancement completed. Enhanced image saved as:", output_image_path)
```
